# Seperating columns into a new dataframe 

```python
new_df = df[['Rescue_ID','Date_TimeCaught','Researcher','CaptureSite','ForagingGround','CaptureMethod','Fisher','LandingSite','Species','TurtleCharacteristics','ReleaseSite','Status']].copy()
```

```python
new_df.info()
```

```python
releasesitemode = new_df['ReleaseSite'].mode()
new_df['ReleaseSite'] = new_df['ReleaseSite'].fillna(releasesitemode)
```

```python
new_df.info()
```

<b>Seperating the datetimecaught column into year month and week<b> 

```python
#splitting the year from date_timecaught
new_df['YearCaught'] = pd.DatetimeIndex(new_df['Date_TimeCaught']).year
#splitting month
new_df['MonthCaught'] = pd.DatetimeIndex(new_df['Date_TimeCaught']).month
#splitting the week
new_df['WeekCaught'] = pd.DatetimeIndex(new_df['Date_TimeCaught']).week
new_df.head()
```
