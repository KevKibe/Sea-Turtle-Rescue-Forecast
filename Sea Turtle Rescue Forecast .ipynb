<b>Seperating the datetimecaught column into year month and week<b> 

```python
#splitting the year from date_timecaught
newer_df['YearCaught'] = pd.DatetimeIndex(newer_df['Date_TimeCaught']).year
#splitting month
newer_df['MonthCaught'] = pd.DatetimeIndex(newer_df['Date_TimeCaught']).month
#splitting the week
newer_df['WeekCaught'] = pd.DatetimeIndex(newer_df['Date_TimeCaught']).week
newer_df.head()
```

```python
newer_df.drop("Rescue_ID", axis=1, inplace=True)

newer_df.drop("ForagingGround", axis=1, inplace=True)
newer_df.drop("YearCaught", axis=1, inplace=True)
newer_df.drop("MonthCaught", axis=1, inplace=True)
newer_df.info()
```

```python
all_train_day = newer_df[['Date_TimeCaught','CaptureSite']].sort_values(by='Date_TimeCaught').copy()
unique_train_day = all_train_day.drop_duplicates('Date_TimeCaught')

CaptureSite = list(newer_df.CaptureSite.unique())
## Find away to optimize this code
unique_train_day

for CapId in CaptureSite:
    unique_train_day[str(CapId)] = 0
    events = all_train_day[all_train_day['CaptureSite'] == CapId]
    dates = events['Date_TimeCaught'].unique()
    
    for date in dates:
        number_of_days = len(events[events.Date_TimeCaught == date ])
        unique_train_day.loc[unique_train_day['Date_TimeCaught'].isin([date]), CapId] = number_of_days
```
